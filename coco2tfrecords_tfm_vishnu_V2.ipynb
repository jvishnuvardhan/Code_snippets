# Converting data from coco to TFRecords

This colab is to convert data from coco json format to tfrecords dataformat. 

Models in Official repository (of model-garden) require models in a TFRecords dataformat. Please check [this resource](https://www.tensorflow.org/tutorials/load_data/tfrecord) to learn more about TFRecords data format.

```python
# clone the model-garden repo as the required data conversion codes are within this model-garden repository
!git clone https://github.com/tensorflow/models.git
```

```python
# Upload data
!unzip '/content/BCCD.v1i.coco.zip'
```

```python
# change directory to vision/data where data conversion tools are available
%cd models/official/vision/data/
```

```python
# CLI command to convert data. (train data)
# Need to provide 
             # 1. image_dir where images are present 
             # 2. object_annotations_file where annotations are listed in json format
             # 3. output_file_prefix where to write output convered TFRecords files 
!python create_coco_tf_record.py --logtostderr \
  --image_dir="/content/train/" \
  --object_annotations_file="/content/train/_annotations.coco.json" \
  --output_file_prefix="/content/output1/" \
  --num_shards=1
```

```python
# CLI command to convert data. (validation data)
!python create_coco_tf_record.py --logtostderr \
  --image_dir="/content/valid/" \
  --object_annotations_file="/content/valid/_annotations.coco.json" \
  --output_file_prefix="/content/output1/" \
  --num_shards=1
```
